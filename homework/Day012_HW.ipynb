### ① 取出今天所有的發文

```python
import requests as rq
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.html'
r = rq.get(url)

soup = BeautifulSoup(r.text, 'html5lib')

for d in soup.find(class_='part_list_2').find_all('h3'):
     print(d.find(class_="date").text, d.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
category = dict()
for d in soup.find(class_='part_list_2').find_all('h3'):
     category[d.find('em').text] = []
for d in soup.find(class_='part_list_2').find_all('h3'):
    for key in category:
        if d.find('em').text == key:
            category[key].append(d.find('a').text.replace(u'\u3000',u' '))
           
for key in category:
    print(key,[k for k in category[key]])
```

### ③ 哪一個類別的文章最多

```python
maxi = 0
for key in category:
    if len(category[key]) > maxi:
        maxi = len(category[key])
        maxi_cat = key
        
print('「{}」類別的文章最多({}篇)'.format(maxi_cat, maxi))
```
