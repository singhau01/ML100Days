### 1. 根據範例使用 selenuim 取出 Hahow 網站中所有課程的資料，並用一個適合的資料結構儲存（Hint: 將所有資料存在一個變數當中取出）

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source

# 取得資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
soup
```

```python
courses= []
for d in soup.find_all(class_="title marg-t-20 marg-b-10"):
    courses.append(d.text)
print(courses)    
```

```python
browser.close()
```

### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
soup
```

```python
# 1. 先取出全部有幾頁
import re
pages = int(soup.find_all('a', href=re.compile('/courses\?page=\d\d'))[-1].text)
pages
```

```python
# 2. 每頁都取出資料放入 courses
import time
courses = []
for i in range(1, pages+1):
    print('Now is Page',i)
    for d in soup.find_all(class_="title marg-t-20 marg-b-10"): # 抓下頁面內容
        courses.append(d.text)
    try:
        search_btn = browser.find_element_by_class_name('rc-pagination-next') # 按下頁
        search_btn.click()
        time.sleep(6)
        html_source = browser.page_source
        soup = BeautifulSoup(browser.page_source,'html.parser')
    except:
        continue
```

```python
# 3. done!
print(courses)
browser.close()
```
