# 未登入用BeautifulSoup嘗試取得資料
* 無法於 https://github.com/ 登入

```python
res = rq.get('https://github.com/new')
soup = BeautifulSoup(res.text, 'lxml')
print(soup.text)
```

```python
import requests
from bs4 import BeautifulSoup

cookies = {
    'logged_in': 'yes',
}

r = requests.get('https://github.com/new', cookies=cookies)
soup = BeautifulSoup(r.text, 'lxml')
soup.title 
```
