## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
url = 'https://www.ettoday.net/news/news-list.htm'
browser.get(url)
```

```python
import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
for d in soup.find(class_='part_list_2').find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[0].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
import datetime 
now = datetime.datetime.now() # 現在時間
two_hour_bf = now + datetime.timedelta(hours=-2) # 取得兩小時datetime模式

for d in soup.find(class_='part_list_2').find_all('h3'):
    date = d.find(class_="date").text
    date_transformed = datetime.datetime.strptime(date, '%Y/%m/%d %H:%M')
    if date_transformed > two_hour_bf:
        print(d.find(class_="date").text, d.find_all('a')[0].text)
```

```python
browser.quit();
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import re

link = soup.find(class_="part_menu_2").a["href"] # 找出link
regex = re.compile('(\d){4}-(\d){2}-(\d){2}') 
link_date = re.search(pattern, link).group(0) # 分析出link日期
link_dt = datetime.datetime.strptime(link_date, '%Y-%m-%d')
three_day_before_dt =  link_dt + datetime.timedelta(days=-3) # 取得三日前日期
three_day_before_date = datetime.date.strftime(three_day_before,'%Y-%m-%d' )

three_day_before_3pm = three_day_before_dt.replace(hour=15,minute=0,second=0,microsecond=0)
three_day_before_5pm = three_day_before_dt.replace(hour=17,minute=0,second=0,microsecond=0)
```

```python
import time
new_url = 'https://www.ettoday.net' + link.replace(link_date, three_day_before_date)
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(new_url)
while True:
    browser.execute_script("window.scrollTo(0, 10000);")
    time.sleep(1)
    new_html_source = browser.page_source
    new_soup = BeautifulSoup(new_html_source, "html5lib")
    end_time = new_soup.find(class_="part_list_2").find_all('h3')[-1].find(class_="date").text
    if datetime.datetime.strptime(end_time, '%Y/%m/%d %H:%M') < three_day_before_3pm:
        print('==== STOP ====')
        break
```

```python
now = datetime.datetime.now()
three_day_before = now + datetime.timedelta(days=-3)
three_day_before_3pm = three_day_before.replace(hour=15,minute=0,second=0,microsecond=0)
three_day_before_5pm = three_day_before.replace(hour=17,minute=0,second=0,microsecond=0)

for d in new_soup.find(class_='part_list_2').find_all('h3'):
    date = d.find(class_="date").text
    date_transformed = datetime.datetime.strptime(date, '%Y/%m/%d %H:%M')
    if date_transformed < three_day_before_5pm and date_transformed > three_day_before_3pm:
        print(d.find(class_="date").text, d.find_all('a')[0].text)
```
