## requests + BeautifulSoup


```python
import requests as rq
from bs4 import BeautifulSoup

url = 'https://money.udn.com/money/index'
headers = headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36'}
r = rq.get(url, headers=headers)
r.text
```

```python
soup = BeautifulSoup(r.text)
print(soup.title.text)
```

## Grab + PyQuery

```python
from grab import Grab
g = Grab()
resp = g.go('https://money.udn.com/money/index')
resp.body
```

```python
from pyquery import PyQuery as py
doc = py(resp.body)
title = doc('title')
print(type(title), title.text())
```
