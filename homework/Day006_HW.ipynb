    1. 取出知乎問題發問時間
    2. 取出第一筆與最後一筆回答的時間


### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間
import requests as rq
import json
import time

headers = {'user-agent': 'my-app/0.0.1'}
url = 'https://www.zhihu.com/api/v4/questions/55493026/answers'

r = rq.get(url, headers=headers)
response = r.text
data = json.loads(response)
num = 0
for d in data['data']:
    num += 1
    print('問題{}發問時間: {}'.format(num, time.ctime(d['created_time'])))
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間
print('第一筆回答時間： {}'.format(time.ctime(data['data'][0]['updated_time'])))
print('最後一筆回答時間： {}'.format(time.ctime(data['data'][-1]['updated_time'])))
```
