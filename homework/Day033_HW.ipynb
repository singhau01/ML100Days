# 登入介面搭配selenium登入
* 由於無法在'https://github.com/login' 直接登入

```python
res_login = rq.get(url)
soup = BeautifulSoup(res_login.text, 'lxml')
print(soup.prettify())
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait

browser = webdriver.Chrome()
browser.maximize_window()  # 最大化視窗
wait = WebDriverWait(browser, 10) # 等待載入10s

def login():
    browser.get(url)
    input = wait.until(EC.presence_of_element_located(
        (By.XPATH, '//*[@id="login_field"]')))
    input.send_keys('帳號') # 輸入帳號
    input = wait.until(EC.presence_of_element_located(
        (By.XPATH, '//*[@id="password"]')))
    input.send_keys('密碼') # 輸入密碼
    submit = wait.until(EC.element_to_be_clickable(
        (By.XPATH, '//*[@id="login"]/div[4]/form/div/input[12]')))
    submit.click() # 點選登入按鈕
    
login()
```

```python
html = browser.page_source
soup = BeautifulSoup(html, 'lxml')
# soup.prettify
```
